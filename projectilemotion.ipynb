```python
#Air Resistance Analytical
velo =  50
alphac = 35*pi/180
grounds = 9.8
tbeg = 0
deltate = 0.01
Di = 0.0013
mass = 0.15
```

```python
a_x0 = -(Di/mass)*Vel0*v_x0
a_y0 = -grounds-(Di/mass)*Vel0*v_y0
a_x0, a_y0
```

```python
ttot = (2*v_y0)/grounds
ttot
```

```python
X_beg = 0
Y_beg = 0
```

```python
X_var = [0]
Y_var = [0]
while tbeg <= ttot:
    xo =  (v_x0 * tbeg)  + (0.5*a_x0*(tbeg**2))
    X_var.append(xo)
    yo =  (v_y0*tbeg) - (0.5*a_y0*(tbeg**2))
    Y_var.append(yo)
    tbeg = tbeg+deltate

X_var,Y_var
```
