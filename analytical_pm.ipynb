```python
import matplotlib.pyplot as plt
from math import *
v = 50
theta = 35*pi/180
g = 9.8
t = 0
dt = 0.01
```

```python
vx = v*cos(theta)
vy = v*sin(theta)
print(vx,vy)
```

```python
Xo = 0
Yo = 0
```

```python
ttol = (2*vy)/g
print(ttol)
```

```python
X = []
Y = []
while t <= ttol:
    x = Xo + vx * t
    X.append(x)
    y = Yo + (vy*t) - ((g*t**2)/2)
    Y.append(y)
    t = t+dt

X,Y
```

```python
plt.plot(X,Y)
plt.show()
```
